```python
#a. Extraer la información del archivo. 
import pandas as pd
df = pd.read_csv('country_vaccinations.csv')
print(df.head())
```

```python
# b. Mostrar la estructura y tipos de datos de cada columna para identificar qué operaciones puedes realizar con cada una de ellas
print(df.info())
# asegurándote que las columnas con fechas sean del tipo datetime64
df['date'] = pd.to_datetime(df['date']) 
```

```python
# c. Determinar la cantidad de vacunas aplicadas de cada compañía 
#(con base en cómo lo reporta cada país en la columna vaccines)
df['company'] = df['vaccines'].str.split(',').str[0]
print(df.company)

vacunas_por_compania = df['company'].value_counts()
print(vacunas_por_compania)
```

```python
#d. Obtener la cantidad de vacunas aplicadas en todo el mundo. 
total_vacunas = df['total_vaccinations'].sum()
print(total_vacunas)
```

```python
# e. Calcular el promedio de vacunas aplicadas por país. 
promedio_vacunas_por_pais = df.groupby('country')['total_vaccinations'].mean()
print(promedio_vacunas_por_pais)
```

```python
# f. Determinar la cantidad de vacunas aplicadas el día 29/01/21 en todo el mundo. 
df_fecha = df.loc[df['date'] == '2021-01-29']
vacunas_el_29 = df_fecha['total_vaccinations'].sum()
print(vacunas_el_29)
```

```python
# g. Crear un dataframe nuevo denominado conDiferencias que contenga los datos originales 
# y una columna derivada (diferencias) con las diferencias de aplicación entre las columnas daily_vaccionations y daily_vaccionations_raw.  
conDiferencias = df.assign(diferencias = df['daily_vaccinations'] - df['daily_vaccinations_raw'])
print(conDiferencias)
```

```python
# h. Obtener el período de tiempo entre el registro con fecha más reciente y el registro con fecha más antigua. 
fecha_minima = df['date'].min()
fecha_maxima = df['date'].max()
periodo_de_tiempo = fecha_maxima - fecha_minima
print(periodo_de_tiempo)
```

```python
#i. Crear un dataframe nuevo denominado conCantidad que contenga los datos originales 
# y una columna derivada (canVac) con la cantidad de vacunas utilizadas cada día (usar la columna vaccines y separar por el carácter /). 
conCantidad = df.assign(canVac = df['vaccines'].str.split('/').apply(lambda x: sum([int(i) for i in x if i.isdigit()])))
print(conCantidad)
```

```python
# j. Generar un dataframe denominado antes20 con todos los registros que se hayan realizado antes del 20 de diciembre de 2020. 
antes20 = df.loc[df['date'] < '2020-12-20']
print(antes20)
```

```python
# k. Obtener un dataframe denominado pfizer con todos los registros donde se haya utilizado la vacuna Pfizer. 
pfizer = df.loc[df['vaccines'].str.contains('Pfizer')]
print(pfizer)
```

```python
# l. Almacenar los dataframes generados (conDiferencias, conCantidad, antes20 y pfizer) en un archivo de Excel denominado resultadosReto.xlsx
# donde cada dataframe ocupe una hoja diferente. 
# Se recomienda ver la documentación de pd.ExcelWriter. 

# creé un objeto ExcelWriter
writer = pd.ExcelWriter('resultadosReto.xlsx', engine='xlsxwriter')

# almacnené cada dataframe en una hoja separada
conDiferencias.to_excel(writer, sheet_name='conDiferencias', index=False)
conCantidad.to_excel(writer, sheet_name='conCantidad', index=False)
antes20.to_excel(writer, sheet_name='antes20', index=False)
pfizer.to_excel(writer, sheet_name='pfizer', index=False)

# guardé el archivo Excel
writer.save()

```
